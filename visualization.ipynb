```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pickle
```

```python
#DANCer

timesteps = [1, 2, 3, 4, 5]

static_S = [0.9145, 0.263, 0.5003, 0.5975, 0.2303]
static_B = [0.4996, 0.4096, 0.4949, 0.5935, 0.5464]
static_T = [0.8656, 0.561, 0.8771, 0.9142, 0.5402]
static_P = [0.8809, 0.5092, 0.8977, 0.831, 0.5473]

chordal_S = [0.9145, 0.4476, 0.5039, 0.6009, 0.2244]
chordal_B = [0.4996, 0.4113, 0.4956, 0.5869, 0.419]
chordal_T = [0.8656, 0.5402, 0.8771, 0.9142, 0.5402]
chordal_P = [0.8809, 0.53, 0.8602, 0.7974, 0.5555]

indicator_S = [0.9145, 0.162, 0.13, 0.3777, 0.514]
indicator_B = []
indicator_T = []
indicator_P = []

plt.plot(timesteps, static_S, color='red', marker='o')
plt.plot(timesteps, chordal_S, color='blue', marker='o')
plt.plot(timesteps, indicator_S, color='green', marker='o')

plt.title('DANCer Community Detection', fontsize=14)
plt.xlabel('Time Step', fontsize=14)
plt.ylabel('NMI', fontsize=14)
red_patch = mpatches.Patch(color='red', label='Static CD')
blue_patch = mpatches.Patch(color='blue', label='Dynamic CD (chordal)')
green_patch = mpatches.Patch(color='green', label='Dynamic CD (indicator)')
plt.legend(handles=[red_patch, blue_patch, green_patch])
plt.xticks(timesteps)
plt.grid(True)

plt.show()
```
